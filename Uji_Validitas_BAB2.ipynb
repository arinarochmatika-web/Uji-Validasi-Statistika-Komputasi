A. Uji Normalitas

```python
#Uji Normalitas (Jarque Bera)
def jarque_bera_test(data):
    n = len(data)

    # Mean
    mean = sum(data) / n

    # Selisih terhadap mean
    selisih = []
    for x in data:
        selisih.append(x - mean)

    # Momen ke-2
    m2 = 0
    for d in selisih:
        m2 += d**2
    m2 /= n

    # Momen ke-3
    m3 = 0
    for d in selisih:
        m3 += d**3
    m3 /= n

    # Momen ke-4
    m4 = 0
    for d in selisih:
        m4 += d**4
    m4 /= n

    # Skewness
    skewness = m3 / (m2 ** 1.5)

    # Kurtosis
    kurtosis = m4 / (m2 ** 2)

    # Jarque-Bera
    jb = (n / 6) * (skewness**2 + ((kurtosis - 3)**2) / 4)

    chi_square_critical = 5.991

    if jb > chi_square_critical:
        keputusan = "Tolak H0: data tidak berdistribusi normal"
    else:
        keputusan = "Gagal menolak H0: data berdistribusi normal"

    print("n =", n)
    print("mean =", mean)
    print("m2 =", m2)
    print("m3 =", m3)
    print("m4 =", m4)
    print("skewness =", skewness)
    print("kurtosis =", kurtosis)
    print("jb =", jb)
    print("keputusan =", keputusan)


data = [12, 15, 14, 10, 18, 16, 13, 17, 11, 14]
jarque_bera_test(data)
```

```python
#Uji Normalitas (Kolmogorov-Smirnov)
from math import erf, sqrt

data = [12, 15, 14, 10, 18, 16, 13, 17, 11, 14]

#Mengurutkan Data
data.sort()

#Menghitung Jumblah Data
n = len(data)

#Menghitung Rata-Rata
mean = sum(data) / n


jumlah_kuadrat = 0
for x in data:
    jumlah_kuadrat += (x - mean) ** 2

std = (jumlah_kuadrat / n) ** 0.5

Fo = []
for i in range(n):
    Fo.append((i + 1) / n)

z_score = []
for x in data:
    z_score.append((x - mean) / std)

Ft = []
for z in z_score:
    Ft.append(0.5 * (1 + erf(z / sqrt(2))))

selisih = []
for i in range(n):
    selisih.append(abs(Fo[i] - Ft[i]))

D_hitung = max(selisih)

D_tabel = 1.36 / (n ** 0.5)

if D_hitung > D_tabel:
    keputusan = "Tolak H0: data tidak berdistribusi normal"
else:
    keputusan = "Gagal menolak H0: data berdistribusi normal"

print("n =", n)
print("mean =", mean)
print("std =", std)
print("D_hitung =", D_hitung)
print("D_tabel =", D_tabel)
print("keputusan =", keputusan)
```

B Uji Homogenitas

```python
#Uji Homogenitas (Uji F Varians)
from scipy.stats import f

def uji_F_varians(data1, data2, alpha=0.05):

    n1 = len(data1)
    n2 = len(data2)

    mean1 = sum(data1)/n1
    mean2 = sum(data2)/n2

    jumlah_kuadrat1 = 0
    for x in data1:
        jumlah_kuadrat1 += (x-mean1)**2

    varians1 = jumlah_kuadrat1/(n1-1)

    jumlah_kuadrat2 = 0
    for x in data2:
        jumlah_kuadrat2 += (x-mean2)**2

    varians2 = jumlah_kuadrat2/(n2-1)

    if varians1 > varians2:
        f_hitung = varians1/varians2
        df1 = n1-1
        df2 = n2-1
    else:
        f_hitung = varians2/varians1
        df1 = n2-1
        df2 = n1-1

    f_tabel = f.ppf(1-alpha, df1, df2)

    if f_hitung > f_tabel:
        keputusan = "Tolak H0: varians tidak homogen"
    else:
        keputusan = "Gagal menolak H0: varians homogen"

    print("n1 =", n1)
    print("n2 =", n2)
    print("mean1 =", mean1)
    print("mean2 =", mean2)
    print("varians1 =", varians1)
    print("varians2 =", varians2)
    print("f_hitung =", f_hitung)
    print("f_tabel =", f_tabel)
    print("keputusan =", keputusan)


data1 = [70,75,80,72,78]
data2 = [65,85,60,90,70]

uji_F_varians(data1,data2)
```

```python
#Uji Homogenitas (Uji Levane)
from scipy.stats import f

data1 = [70, 75, 80, 72, 78]
data2 = [65, 85, 60, 90, 70]

mean1 = sum(data1)/len(data1)
mean2 = sum(data2)/len(data2)

z1 = [abs(x-mean1) for x in data1]
z2 = [abs(x-mean2) for x in data2]

zbar1 = sum(z1)/len(z1)
zbar2 = sum(z2)/len(z2)

gabung = z1 + z2

zbar = sum(gabung)/len(gabung)

N = len(gabung)
k = 2

pembilang = len(z1)*(zbar1-zbar)**2 + \
            len(z2)*(zbar2-zbar)**2

penyebut = 0

for z in z1:
    penyebut += (z-zbar1)**2

for z in z2:
    penyebut += (z-zbar2)**2

W = ((N-k)/(k-1))*(pembilang/penyebut)

f_tabel = f.ppf(0.95,k-1,N-k)

if W > f_tabel:
    keputusan = "Tolak H0: varians tidak homogen"
else:
    keputusan = "Gagal menolak H0: varians homogen"

print("W =", W)
print("F tabel =", f_tabel)
print("Keputusan =", keputusan)
```

C. Uji Independensi

```python
# Uji Independensi Chi-Square
from scipy.stats import chi2

def uji_chi_square(data, alpha=0.05):

    jumlah_baris = len(data)
    jumlah_kolom = len(data[0])

    total_baris = []
    for baris in data:
        total_baris.append(sum(baris))

    total_kolom = []
    for j in range(jumlah_kolom):
        total = 0
        for i in range(jumlah_baris):
            total += data[i][j]
        total_kolom.append(total)

    grand_total = sum(total_baris)

    harapan = []
    for i in range(jumlah_baris):
        baris = []
        for j in range(jumlah_kolom):
            eij = total_baris[i] * total_kolom[j] / grand_total
            baris.append(eij)
        harapan.append(baris)

    chi_square = 0
    for i in range(jumlah_baris):
        for j in range(jumlah_kolom):
            O = data[i][j]
            E = harapan[i][j]
            chi_square += ((O-E)**2)/E

    df = (jumlah_baris-1)*(jumlah_kolom-1)

    chi_tabel = chi2.ppf(1-alpha, df)

    if chi_square > chi_tabel:
        keputusan = "Tolak H0: terdapat hubungan"
    else:
        keputusan = "Gagal menolak H0: tidak terdapat hubungan"

    print("jumlah_baris =", jumlah_baris)
    print("jumlah_kolom =", jumlah_kolom)
    print("grand_total =", grand_total)
    print("harapan =", harapan)
    print("chi_square =", chi_square)
    print("chi_tabel =", chi_tabel)
    print("keputusan =", keputusan)


data = [[30,20],
        [15,35]]

uji_chi_square(data)
```

```python
# Uji Independensi Fisher's Exact Test
from math import factorial

data = [[30,20],
        [15,35]]

a = data[0][0]
b = data[0][1]
c = data[1][0]
d = data[1][1]

n = a+b+c+d

p = (
    factorial(a+b) *
    factorial(c+d) *
    factorial(a+c) *
    factorial(b+d)
) / (
    factorial(a) *
    factorial(b) *
    factorial(c) *
    factorial(d) *
    factorial(n)
)

alpha = 0.05

if p < alpha:
    keputusan = "Tolak H0: terdapat hubungan"
else:
    keputusan = "Gagal menolak H0: tidak terdapat hubungan"

print("a =", a)
print("b =", b)
print("c =", c)
print("d =", d)
print("n =", n)
print("p-value =", p)
print("keputusan =", keputusan)
```
